# EDA

```python
#Univarite Analysis
import seaborn as sn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sn.distplot(df["Malic acid"])
plt.title("Acid data")
plt.show()
```

```python
sn.countplot(df["Class"])
plt.title("Different Class")
plt.show()
```

```python
plt.scatter(df["Ash"],df["Malic acid"])
plt.title("Scatterplot")
plt.xlabel("Ash")
plt.ylabel("Malic acid")
plt.show()
```

```python
plt.violinplot(df["Color intensity"])
```

```python
#Checking correlation with heatmap
corr=df.corr()
sn.heatmap(corr,annot=True)
plt.show()
```

```python
sn.boxplot(data=df[["Ash","Malic acid"]])
plt.title("Outliers Detection using Boxplot")
plt.show()
```

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
threshold=2
new_df=df[(z < 3).all(axis=1)]
df.shape
```
